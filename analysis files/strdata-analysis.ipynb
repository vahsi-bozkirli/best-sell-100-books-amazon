```python
def return_lenuniq(data:pd.DataFrame,col:str):
    """
    data: required pandas DataFrame from data in prepared_data
    col: data type is str columns from data

    this function return length and unique values of col parameter

    example usage: return_lenuniq(top_books,"genre")
    """  
    arr = data[col].unique()
    
    return {
        "length of {} array:".format(col):len(arr),
        "unique array:":arr
    }

return_lenuniq(top_books,"genre")
```

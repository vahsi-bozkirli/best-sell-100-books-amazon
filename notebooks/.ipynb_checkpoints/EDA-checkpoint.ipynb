```python
datasets = top_books.select_dtypes(["int","float"]).drop("year of publication",axis=1)

fig, axs = plt.subplots(1, 3, figsize=(15, 5));

sbn.boxplot(datasets["book price"],ax=axs[0]);
sbn.boxplot(datasets["count genre"],ax=axs[1]);
sbn.boxplot(datasets["rating"],ax=axs[2]);
```

```python
datasets = top_books.select_dtypes(["int","float"]).drop("year of publication",axis=1)

fig, axs = plt.subplots(1, 3, figsize=(15, 5));

sbn.distplot(datasets["book price"],ax=axs[0]);
sbn.distplot(datasets["count genre"],ax=axs[1]);
sbn.distplot(datasets["rating"],ax=axs[2]);
```

```python
sbn.pairplot(top_books,corner=True,markers="+",diag_kind="kde");
```

### Notlar
- fiyat ve rating arasında bir ilişki yok fiyatı düşük olan kitaplar daha çok inceleme almış ve fiyatı yüksek olan kitaplar daha az inceleme almış (bu veri setine göre)
- fiyatı düşük olan kitaplarda, fiyatı düşük olan kitaplarda ratingleri ortak olarak yüksek ancak fiyatı düşük olan kitaplarda düşük ratingli kitaplarda var

- yıllar geçtikçe fiyatlar artma eğilimi göstermiş veya geçmiş yıllarda değerlendirilen kitap sayısı az
- yıllar geçtikçe ratingler düşme eğilimi göstermiş veya geçmiş yıllarda değerlendirilen kitap sayısı az

- fiyat ve tür çeşitliliği arasında bir ilişki yok
- tür çeşitliliği ve rating arasında bir ilişki yok

- yıllar geçtikçe tür çeşitliliği artmış veya (gerisini biliyorsunuz)

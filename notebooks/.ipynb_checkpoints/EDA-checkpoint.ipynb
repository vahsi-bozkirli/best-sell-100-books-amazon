```python
datasets = top_books.select_dtypes(["int","float"]).drop("year of publication",axis=1)

fig, axs = plt.subplots(1, 3, figsize=(15, 5));

sbn.boxplot(datasets["book price"],ax=axs[0]);
sbn.boxplot(datasets["count genre"],ax=axs[1]);
sbn.boxplot(datasets["rating"],ax=axs[2]);
```

```python
datasets = top_books.select_dtypes(["int","float"]).drop("year of publication",axis=1)

fig, axs = plt.subplots(1, 3, figsize=(15, 5));

sbn.distplot(datasets["book price"],ax=axs[0]);
sbn.distplot(datasets["count genre"],ax=axs[1]);
sbn.distplot(datasets["rating"],ax=axs[2]);
```

```python
sbn.pairplot(top_books,corner=True,markers="+",diag_kind="kde");
```

### Notlar
- fiyat ve rating arasnda bir iliki yok fiyat dk olan kitaplar daha ok inceleme alm ve fiyat yksek olan kitaplar daha az inceleme alm (bu veri setine gre)
- fiyat dk olan kitaplarda, fiyat dk olan kitaplarda ratingleri ortak olarak yksek ancak fiyat dk olan kitaplarda dk ratingli kitaplarda var

- yllar getike fiyatlar artma eilimi gstermi veya gemi yllarda deerlendirilen kitap says az
- yllar getike ratingler dme eilimi gstermi veya gemi yllarda deerlendirilen kitap says az

- fiyat ve tr eitlilii arasnda bir iliki yok
- tr eitlilii ve rating arasnda bir iliki yok

- yllar getike tr eitlilii artm veya (gerisini biliyorsunuz)

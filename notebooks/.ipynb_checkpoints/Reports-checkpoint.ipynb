```python
import pandas as pd
import numpy as np
import nltk
import matplotlib.pyplot as plt
import seaborn as sbn
import pandasql as psql
from read_data import import_py_file
from warnings import filterwarnings
from sklearn.preprocessing import MinMaxScaler
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
filterwarnings("ignore")

nltk.download('vader_lexicon')
analyzer = SentimentIntensityAnalyzer()
mms = MinMaxScaler()

path = '../'
import_py_file(path)

import preprocessed_file as pf
```

```python
read_data = pf.Data("../data/raw/Top-100 Trending Books.csv",
                    "../data/raw/customer reviews.csv")
```

```python
def get_season(month):
    if 3 <= month <= 5:
        return "Spring"
    elif 6 <= month <= 8:
        return "Summer"
    elif 9 <= month <= 11:
        return "Autumn"
    else:
        return "Winter"

def get_quarter(month):
    if 1 <= month <= 3:
        return 1
    elif 4 <= month <= 6:
        return 2
    elif 7 <= month <= 9:
        return 3
    else:
        return 4
```

```python
def duygu_analizi(text):
    sentiment = analyzer.polarity_scores(text)
    if sentiment['compound'] >= 0.05:
        return 'Pozitif'
    elif sentiment['compound'] <= -0.05:
        return 'Negatif'
    else:
        return 'Nötr'
```

```python
top_books, custrew = read_data.added_features()
```

```python
top_books = top_books.drop(["url","link","book","dp","ref","slug"],axis=1)
custrew = custrew.drop(["is_verified","timestamp"],axis=1)
```

```python
data = psql.sqldf("select * from custrew c inner join top_books t on t.id=c.ASIN")
data = data.drop(["id","book title"],axis=1)
```

```python
data[["day","month","year"]] = data["date"].str.split("-",expand=True)
data[["day","month","year"]] = data[["day","month","year"]].astype("int")
```

```python
data['season'] = data['month'].apply(get_season)
data['quarter'] = data['month'].apply(get_quarter)

data["title sentiment result"] = data["review title"].apply(duygu_analizi)
data["desc sentiment result"] = data["review description"].apply(duygu_analizi)
```
